```python
import csv
```

```python
f = open('30oysters.csv', 'rb')
reader = csv.reader(f)
rownum = 0
weight = []
vol = []
for row in reader:
    if rownum == 0:
        header = row
    else:
        r = map(float, row[0].split())
        weight.append(r[1])
        vol.append(r[2])
    rownum += 1
f.close()
```
